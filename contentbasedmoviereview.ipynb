```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
df=pd.read_csv("movie_dataset.csv")
df.head()
```

Replacing nan values with emptystrings


```
features=["keywords","cast","genres","director"]
for feature in features:
  df[feature]=df[feature].fillna('')
```

To combine required features into a single column(feature)

```
def combine_features(row):
  return row["keywords"]+" "+row["cast"]+" "+row["genres"]+" "+row["director"]
df["combined_features"]=df.apply(combine_features,axis=1)
```

to check combined feature


```
df.iloc[0]["combined_features"]
```

To transform into a countvector and perform cosine similarity


```
cv=CountVectorizer()
count_matrix=cv.fit_transform(df["combined_features"])
cosine_model=cosine_similarity(count_matrix)
cosine_model_db=pd.DataFrame(cosine_model,index=df.title,columns=df.title)
cosine_model_db.head()
```

Make recommendation

```
def make_recommendation(movie_name):
  return cosine_model_db[movie_name].sort_values(ascending=False)
make_recommendation("Avatar")
```
